```python
auto=pd.read_csv("titanic.csv")
auto.head()
```

```python
sexvssurvived=auto.groupby(["sex","survived"]).size().unstack(fill_value=0)
sexvssurvived
```

```python
sexvssurvived.plot(kind='bar',stacked=True)
plt.show()
```

```python
data = pd.read_csv('titanic.csv')
plt.figure(figsize=(8, 6))
sns.histplot(data['age'].dropna(), bins=30, kde=True, color='red')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
survival_rate_by_class = data.groupby('pclass')['survived'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='pclass', y='survived', data=survival_rate_by_class, palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='embark_town', data=data, palette='muted')
plt.title('Number of Passengers by Embarkation Town')
plt.xlabel('Embarkation Town')
plt.ylabel('Count')
plt.show()
```

```python
selected_columns = ['age', 'fare', 'pclass', 'survived']
subset_data = data[selected_columns].dropna()

sns.pairplot(subset_data, hue='survived', palette='coolwarm')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```
